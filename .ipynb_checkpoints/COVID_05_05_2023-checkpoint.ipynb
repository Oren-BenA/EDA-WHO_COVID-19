```python
from datetime import datetime

fname = r'C:\Users\user\Documents\Oren\Data Science_NAYA_02-2023\My first Python Project_COVID\git_COVID-19_project\WHO-COVID-19-global-data.csv'

g_table_fname = r'C:\Users\user\Documents\Oren\Data Science_NAYA_02-2023\My first Python Project_COVID\git_COVID-19_project\WHO-COVID-19-global-table-data.csv'

class Covid:

  def __init__(self,file_name, glbl_dct, g_table_fname, g_table_dct, g_monthly_dct):
   self.file_name = file_name
   self.glbl_dct = glbl_dct = self.make_dct(file_name)
   self.g_table_fname = g_table_fname 
   self.g_table_dct = g_table_dct = self.g_table_data_to_dct(g_table_fname)
   self.global_monthly_dict = g_monthly_dct = self.global_monthly_dict(file_name)

  
  def headers_file(self, file_name):
    with open(file_name,'r') as f:
      lines = f.read().split('\n')
      headers = lines[0].split(',')
    return headers

  def headers_g_table(self, g_table_fname):
    with open(g_table_fname, 'r') as gt:
      lines = gt.read().split('\n')
      gt_headers = lines[0].split(',')
      return gt_headers
    
  def make_dct(self, file_name):
    with open(fname,'r') as f:
      f_content = f.read().split('\n')
    glbl_dct = {}

    for line in f_content[1:]:
      if len(line.split(',')) == 8:
        country = line.split(',')[2]
      if country not in glbl_dct.keys():
        glbl_dct[country] = {}
      if len(line.split(',')) == 8:
        s_date = line.split(',')[0]
        date = datetime.strptime(s_date,'%Y-%m-%d')
        date = date.date()
        new_cases, cumulative_cases, new_deaths,cumulative_deaths = line.split(',')[4],line.split(',')[5],line.split(',')[6],line.split(',')[7]
        glbl_dct[country][date] = (int(new_cases),int(cumulative_cases), int(new_deaths),int(cumulative_deaths))
    return glbl_dct

  def first_case_date_per_country(self,glbl_dct):
    countries = list(glbl_dct.keys())
    first_case_date_in_country = []
    for country in countries:
      for date, tup in glbl_dct[country].items():
        if tup[0] == 0:
          continue
        else:
          first_case_date_in_country.append((country, date))
          break
    return sorted(first_case_date_in_country, key = lambda x : x[1])

  def calc_local_maxes_mins(self, glbl_dct, country):
    week_total = 0
    i=0
    week_daily_mean = []
    weekly_peaks = []
    weekly_mins = []
    flag = True
    
# converting the daily new cases data into a week-based average new cases daily data  
    for key in glbl_dct[country].keys():
      week_total += glbl_dct[country][key][0]
      i+=1
      if i==7:
        week_daily_mean.append((f'{key}', round(week_total/7,1)))
        i=0
        week_total = 0
    
    for j in range(0,len(week_daily_mean)-2):
      if flag:
        # find local maxes
        if week_daily_mean[j][1] <= week_daily_mean[j+1][1]:
          continue
        elif week_daily_mean[j][1] > week_daily_mean[j+2][1]:
          weekly_peaks.append(week_daily_mean[j])
          flag = False
          continue
      if not flag:
        # find local mins
        if week_daily_mean[j][1] >= week_daily_mean[j+1][1]:
          continue
        elif week_daily_mean[j][1] < week_daily_mean[j+2][1]:
          weekly_mins.append(week_daily_mean[j])
          flag = True
    
    return country, weekly_peaks, weekly_mins

  def g_table_data_to_dct(self,g_table_fname):
    output_headers = "[('Country name',(cases_cumulative_total_per_100k, new_cases_last_7_days_per_100k, cases_cumulative_deaths_per_100k,new_deaths_last_7_days_per_100k))]"
# for output example see: [('Tuvalu', (23566.8, 0.0, 0.0, 0.0)),
                                      
    g_table_data = ()
    g_table_dct = {}
    country = ''
    cases_cumul_total_per_100k = 0
    cases_last_7_days_per_100k = 0
    cumul_deaths_per_100k = 0
    deaths_last_7_days_per_100k = 0
    sorted_g_table_dct = {}
    enum_sorted_g_table_dct = []

    with open(g_table_fname, 'r') as gt:
      g_table_data = gt.read().split('\n')
    for line in g_table_data[2:]:
      if len(line.split(',')) == 12:
        country = line.split(',')[0]
        if len(line.split(',')[3]) > 0:
          cases_cumul_total_per_100k = round(float(line.split(',')[3]),1)
        if len(line.split(',')[5]) > 0:
          cases_last_7_days_per_100k = round(float(line.split(',')[5]),1)
        if len(line.split(',')[8]) > 0:
          cumul_deaths_per_100k = round(float(line.split(',')[8]),1)
        if len(line.split(',')[10]) > 0:
          deaths_last_7_days_per_100k = round(float(line.split(',')[10]),1)
        
        if country not in g_table_dct.keys():
          g_table_dct[country] = cases_cumul_total_per_100k, cases_last_7_days_per_100k,cumul_deaths_per_100k,deaths_last_7_days_per_100k

    return  output_headers, g_table_dct

  def global_monthly_dict(self,file_name):
    g_monthly_dct = {}
    g_m_new_cases, g_m_new_deaths = 0, 0
    i=0
    with open(fname, 'r') as f:
      lines_lst = f.read().split('\n')

    data_lst = []
    for i in range(1,len(lines_lst)-1):
      if len(lines_lst[i].split(',')) == 8:
        date = datetime.strptime(lines_lst[i].split(',')[0],'%Y-%m-%d').date()
        new_cases = int(lines_lst[i].split(',')[4])
        new_deaths = int(lines_lst[i].split(',')[6])
        data_lst.append((date,new_cases,new_deaths))
    sorted_data_lst = sorted(data_lst)

    for i in range(len(sorted_data_lst)-1):
      if sorted_data_lst[i][0].month == sorted_data_lst[i+1][0].month:
        g_m_new_cases = sorted_data_lst[i][1] + sorted_data_lst[i+1][1]
        g_m_new_deaths = sorted_data_lst[i][2] + sorted_data_lst[i+1][2]
      if sorted_data_lst[i][0].month != sorted_data_lst[i+1][0].month:
        g_monthly_dct[(f'{sorted_data_lst[i][0].month}-{sorted_data_lst[i][0].year}')] = (g_m_new_cases, g_m_new_deaths)
    return g_monthly_dct    
```
